```python
output_folder = "/data/hanchong/maritime-vessel-dataset-infrared/Analysis"

for path in paths:
    try:
        a, b, c, d, e, f = path

        category_id = a.split("/")[-2]
        category_name = category_mapping[category_id]

        fig, axs = plt.subplots(2, 3, figsize=(15, 10))
        axs[0, 0].imshow(np.array(Image.open(a).resize((256, 256))))
        axs[0, 0].axis("off")
        axs[0, 0].set_title(f"{category_name} RGB Image")
        axs[1, 0].imshow(np.array(Image.open(b).resize((256, 256))), cmap="gray")
        axs[1, 0].axis("off")
        axs[1, 0].set_title("Segmentation Mask")
        axs[1, 1].imshow(np.array(Image.open(c).resize((256, 256))))
        axs[1, 1].axis("off")
        axs[1, 1].set_title("FLIR (RGB + Mask -> IR)")
        axs[1, 2].imshow(np.array(Image.open(d).resize((256, 256))))
        axs[1, 2].axis("off")
        axs[1, 2].set_title("M3FD (RGB + Mask -> IR)")
        axs[0, 1].imshow(np.array(Image.open(e).resize((256, 256))))
        axs[0, 1].axis("off")
        axs[0, 1].set_title("FLIR (RGB -> IR)")
        axs[0, 2].imshow(np.array(Image.open(f).resize((256, 256))))
        axs[0, 2].axis("off")
        axs[0, 2].set_title("M3FD (RGB -> IR)")

        fig.savefig(f"{output_folder}/{category_id}_{category_name.replace(" ", "")}.png", dpi=300, bbox_inches='tight')
        plt.close(fig)

    except Exception as e:
        print(f"Error processing: {e}")
```
